# Variables
#### elements in programming languages that are used to assign and store data so that it can be manipulated in a program later. (can be used or referenced later)

```python
# Declaring and assigning variables

age=20
height= 5.9
name="Siddhesh"
is_student=True

# printing

print("age:",age)
print("Height:",height)
print("Name:",name)
print("Is student:",is_student)
```

```python
# Naming conventions

## variable names shoulld be descriptive (or use comments to describe)
## they must start with a letter or underscore, and acn contain numbers, but not start with them.
## variable names are case-sensitive

# some valid variable names
first_name="Siddhesh"
last_name="Khankhoje"

# invalid variable names
#1st_name="Siddhesh" starts with a number
```

```python
# Variable types
# Python is Dynamically typed, meaning type of a vriable is determined at runtime
age=20
height=5.9
name="Siddhesh"
is_student=True

print(type(age))
print(type(height))
print(type(name))
print(type(is_student))
```

```python
# Data type conversion
age=20
print(type(age))

age_str=str(age)
print(age_str)
print(type(age_str))
```

```python
age='20'
age_int=int(age)
print(age_int)
print(type(age_int))

#Note: Converting non-numeric strings to int or float will raise an error
```

```python
height=5.9
type(height)

int(height) #removes decimal
float(int(height))
```

```python
# Dynamic typing
## Python allows the type of a variable to change as the program executes
var=10
print(var,type(var))

var="String"
print(var,type(var))

var=3.14
print(var,type(var))
```
