## Basics and  Syntax in Python
'#' is used to write single line comments in python
''' is used for multi line comments. 
### 1. Case Sensitivity - Python is case sensitive

```python
# if we write two different variables or functions but with the same name, different spellings, python interprets them as different entities.

name="Siddhesh"
Name="Khankhoje"

print(name)
print(Name)

# here, the variable 'name' and 'Name' are different.
```

## 2. Indentation
#### Used to define blocks of code, 4 spaces or a tab is required for indentation

```python
age=20
if age>18:
    print("You are an adult.")
```
